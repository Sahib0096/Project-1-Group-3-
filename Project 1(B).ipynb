```python
# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import os
import matplotlib.pyplot as plt
```

```python
household_df = pd.read_csv(r"C:\Users\nikhi\OneDrive\Documents\Fintech\Project 1\canada_household_econimic_data.csv", index_col = 'Income, consumption and savings')
display(household_df.head())
display(household_df.tail())
```

```python
household_df.info()
```

```python
# Selecting Household Consumption and Net Savings Data from full data for Analysis

consumption_df = household_df.transpose().drop(columns=['Household disposable income',
                                                       'Compensation of employees',
                                                       'Net mixed income', 'Property income received',
                                                       'Less: property income paid',
                                                       'Current transfers received',
                                                       'Less: current transfers paid',
                                                       'Social transfers in kind (STiK)',
                                                       'Social transfers in kind (STiK): education',
                                                        'Social transfers in kind (STiK): health',
                                                        'Social transfers in kind (STiK): other',
                                                       'Adjusted household disposable income',
                                                        'Household final consumption expenditure (HFCE)',
                                                       'Household actual final consumption',
                                                       'Change in pension entitlements'], axis=1)


consumption_df
```

```python

# Define a function to remove commas and convert to float
def convert_to_float(value):
    return float(str(value).replace(',', ''))

# Apply conversion function to each specified column
for column in consumption_df.columns:
    consumption_df[column] = consumption_df[column].apply(convert_to_float)

# Print the DataFrame with the updated columns
display(consumption_df.head())

# Print the data types of the columns
consumption_df.dtypes
```

```python
# New Columns with consumption data Sector-wise
consumption_df['Consumer_Staples'] = consumption_df['Food and non-alcoholic beverages'] + consumption_df['Alcoholic beverages and tobacco'] + consumption_df['Clothing and footwear']

consumption_df['Real_Estate and Utilities'] = consumption_df['Housing, water, electricity, gas and other fuels']+consumption_df['Furnishings, household equipment and other goods and services related to the dwelling and property']+consumption_df['Miscellaneous goods and services']

consumption_df['Travel and Leisure'] = consumption_df['Transport']+consumption_df['Recreation and culture']

sector_consumption_df = consumption_df[['Consumer_Staples','Real_Estate and Utilities','Health',
                                       'Insurance and financial services','Travel and Leisure',
                                       'Household net saving']].reset_index()

sector_consumption_df.rename(columns={'index':'Year'},inplace=True)

sector_consumption_df = sector_consumption_df.set_index('Year')

sector_consumption_df

sector_consumption_df['Total_Consumption'] = sector_consumption_df['Consumer_Staples']+sector_consumption_df['Real_Estate and Utilities']+sector_consumption_df['Health']+sector_consumption_df['Insurance and financial services']+sector_consumption_df['Travel and Leisure']

# Print the Sector-wise DataFrame
sector_consumption_df
```

```python
# Change the data into percent change
# Comparing it with Inflation CPI_Common
# Plotting the changes _ using SNS, HVPLOT
```

```python
change_df = sector_consumption_df.pct_change() * 100
change_df
```

```python
new_df = sector_consumption_df.drop(columns=['Consumer_Staples',
                                                       'Real_Estate and Utilities',
                                                       'Health', 'Insurance and financial services',
                                                       'Travel and Leisure'], axis=1)

new_df
```

```python
new_df = pd.DataFrame(new_df)
# Calculate 'Year on Year Increase on Total Consumption'
new_df['Year on Year Increase on Total Consumption'] = new_df['Total_Consumption'].diff()

# Calculate 'Year on Year Increase as a Percentage of Net Savings'
new_df['Year on Year Increase as a Percentage of Net Savings'] = (
    (new_df['Year on Year Increase on Total Consumption'] / new_df['Household net saving'])*100
)
new_df
```

```python
# Quarterly CPI Data

cpi_path = (r"C:\Users\nikhi\OneDrive\Documents\Fintech\Project 1\canada_cpi.csv")
cpi_df = pd.read_csv( cpi_path, skiprows=1)

cpi_df.head()
```

```python
#formating dates by removing Q1, Q2, Q3 and Q4

cpi_df["date"] = [date.split('Q')[0] for date in cpi_df["date"]]

#displaying the first five observations
cpi_df.head()
```

```python
# changing column names
cpi_df.rename(columns = {'date':'Year','INDINF_CPI_MEDIAN_Q':'CPI_MEDIAN','INDINF_CPI_Q':'CPI_CORE',
                               'INDINF_CPI_TRIM_Q':'CPI_TRIM','INDINF_CPI_COMMON_Q':'CPI_COMMON'}, 
                     inplace = True)

# calculating approximate annual inflation as Mean of Quarterly CPI

cpi_annual_df = cpi_df.groupby('Year').mean()

cpi_annual_df
```

```python
import hvplot.pandas

# Slicing cpi_annual_df to choose data for 2011 to 2023

cpi_annual_filetered_df = cpi_annual_df.loc['2011':'2023']

CPI_plot = cpi_annual_filetered_df[['CPI_COMMON']].hvplot()
```
